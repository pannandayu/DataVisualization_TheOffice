```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
"""
1. Create a matplotlib scatter plot of the data that contains the following attributes:
    Each episode's episode number plotted along the x-axis
    Each episode's viewership (in millions) plotted along the y-axis
        A color scheme reflecting the scaled ratings (not the regular ratings) of each episode, such that:
            a. Ratings < 0.25 are colored "red"
            b. Ratings >= 0.25 and < 0.50 are colored "orange"
            c. Ratings >= 0.50 and < 0.75 are colored "lightgreen"
            d. Ratings >= 0.75 are colored "darkgreen"
        A sizing system, such that episodes with guest appearances have a marker size of 250 and episodes without are sized 25
        A title, reading "Popularity, Quality, and Guest Appearances on the Office"
        An x-axis label reading "Episode Number"
        A y-axis label reading "Viewership (Millions)"

2. Provide the name of one of the guest stars (hint, there were multiple!) who was in the most watched Office episode. 
  Save it as a string in the variable top_star (e.g. top_star = "Will Ferrell").
"""
```

```python
df = pd.read_csv('the_office_series.csv')
df.rename(columns=({'Unnamed: 0':'Episode'}), inplace=True)

df['Date'] = pd.to_datetime(df['Date'])
df['GuestStars'] = df['GuestStars'].astype(str)

df.head()
```

```python
df.describe()
```

```python
new_rating = []
guest = []

for index, rows in df.iterrows():
    if rows['Ratings'] < 7.8:
        new_rating.append('red')
    elif (rows['Ratings'] < 8.2):
        new_rating.append('orange')
    elif rows['Ratings'] < 8.6:
        new_rating.append('lightgreen')
    else:
        new_rating.append('darkgreen')

for index, rows in df.iterrows():
    if rows['GuestStars'] == 'nan':
        guest.append(25)
    else:
        guest.append(250)
```

```python
df['colors'] = new_rating
df['size'] = guest
```

```python
df.head(1)
```

```python
df[df['Viewership'] == 22.91]
```

```python
fig, ax = plt.subplots(figsize=(11,7), frameon=True)
plt.style.use('classic')
plt.grid(True)
plt.scatter(x=df['Episode'], y=df['Viewership'], c=df['colors'], s=df['size'], label='Small = no guest\nBig = have guest(s)\n\
\nLG/DG = mid-high rating\nR/O = mid-low rating')

plt.annotate('Cloris Leachman,\nJack Black,\nand Jessica Alba\n(as guest stars)', [53,19])
plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.legend()

plt.show()
```
